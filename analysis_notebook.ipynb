```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import streamlit as st
import plotly.graph_objects as go
```

```python
df = pd.read_csv('retail_sales.csv')
```

```python
df
```

```python
df.head()
```

```python
df.info()
```

```python
# removed duplicates

df.duplicated().sum()
```

```python
# standardize all date formats

df['OrderDate'] = pd.to_datetime(df['OrderDate'],errors='coerce')
```

```python
df
```

```python
#converted money cols into numbers

money_cols = ['UnitPrice', 'Sales', 'Cost','Profit','ProfitMargin']
for col in money_cols:
    df[col] = df[col].replace('[\$,]', '', regex=True).astype(float)
    
```

```python
#fixing category types

cols = df.select_dtypes(include='object').columns
for col in cols:
    df[col] = df[col].str.strip().str.title()
```

```python
df
```

```python
#standerdized returned flag

df['Returned'] = df['Returned'].astype(str).str.strip().str.title()
df['Returned'] =df['Returned'].replace({
    'Y':'Yes',
    'YES':'Yes',
    'N':'No',
    'NO':'No'})
```

```python
df.isnull().sum()
```

```python
#handle missing values

df['PromoCode'] = df['PromoCode'].fillna('No Promocode').inplace=True
```

```python
df.isnull().sum()
```

```python
#removing invalid quantites
df = df[df['Quantity'] > 0]
```

```python
df.to_csv('cleaned_retail_sales.csv', index=False)
```

monthly sales trend (seasonality)

```python
# Monthly Sales trend analysis(seasonality)

df['YearMonth'] = df['OrderDate'].dt.to_period('M')

#monthly sales 
monthly_sales = df.groupby('YearMonth')['Sales'].sum()

#plot
plt.figure(figsize=(12,6))
monthly_sales.plot(kind='line', marker='o',linewidth=2)
plt.title('Monthly Sales Trend(2021-2024)')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

Sales and Profit by region

```python
#sales and profit by region

from matplotlib.ticker import FuncFormatter


region_sales_profit = (
    df.groupby('Region')[['Sales', 'Profit']]
    .sum()
    .reset_index()
)

x = np.arange(len(region_sales_profit['Region']))
width = 0.25  # thinner bars

plt.figure(figsize=(10,6))

plt.bar(
    x - width/2,
    region_sales_profit['Sales'],
    width,
    label='Sales'
)

plt.bar(
    x + width/2,
    region_sales_profit['Profit'],
    width,
    label='Profit'
)

plt.title('Total Sales and Profit by Region')
plt.xlabel('Region')
plt.ylabel('Amount')

plt.xticks(x, region_sales_profit['Region'], rotation=45)
# Convert Y-axis labels to millions
ax = plt.gca()
ax.yaxis.set_major_formatter(
    FuncFormatter(lambda y, _: f'{y/1_000_000:.1f}M')
)

plt.legend()

plt.tight_layout()
plt.show()
```

Sales by category

```python
category_sales = (
    df.groupby('Category')['Sales']
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(10,6))

category_sales.plot(kind='bar')

plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
# Convert Y-axis labels to millions
ax = plt.gca()
ax.yaxis.set_major_formatter(
    FuncFormatter(lambda y, _: f'{y/1_000_000:.1f}M')
)

plt.tight_layout()
plt.show()
```

Year over Year growth

```python

df['Year'] = df['OrderDate'].dt.year

yearly_sales = (
    df.groupby('Year')['Sales']
    .sum()
)

# Calculate YoY Growth
growth = yearly_sales.pct_change() * 100

# Replace first NaN with 0
growth = growth.fillna(0)

plt.figure(figsize=(10,6))

plt.plot(
    growth.index,
    growth.values,
    marker='o',
    markersize=10,
    linewidth=3,
    color='#1f77b4'
)

plt.fill_between(
    growth.index,
    growth.values,
    alpha=0.15,
    color='#1f77b4'
)

# Labels
for year, value in zip(growth.index, growth.values):
    plt.annotate(
        f'{value:.1f}%',
        (year, value),
        textcoords="offset points",
        xytext=(0,10),
        ha='center',
        fontsize=10,
        fontweight='bold'
    )

plt.title(
    'Year-over-Year Sales Growth (%)',
    fontsize=18,
    fontweight='bold'
)

plt.xlabel('Year')
plt.ylabel('Growth (%)')

plt.grid(True, linestyle='--', alpha=0.4)

plt.xticks(growth.index)

plt.tight_layout()
plt.show()
```

calculate KPI

```python
# Total Revenue
total_revenue = df['Sales'].sum()

# Total Profit
total_profit = df['Profit'].sum()

# Profit Margin
profit_margin = (total_profit / total_revenue) * 100

print("=" * 40)
print(f"Total Revenue : ${total_revenue:,.2f}")
print(f"Total Profit  : ${total_profit:,.2f}")
print(f"Profit Margin : {profit_margin:.2f}%")
print("=" * 40)
```

```python
# KPI Calculations
total_revenue = df['Sales'].sum()
total_profit = df['Profit'].sum()
profit_margin = (total_profit / total_revenue) * 100
total_orders = df['OrderID'].nunique()

fig = go.Figure()

fig.add_trace(go.Indicator(
    mode="number",
    value=total_revenue,
    title={"text": "Revenue"},
    number={'prefix': '$', 'valueformat': '.2s'},
    domain={'x': [0.00, 0.23], 'y': [0, 1]}
))

fig.add_trace(go.Indicator(
    mode="number",
    value=total_profit,
    title={"text": "Profit"},
    number={'prefix': '$', 'valueformat': '.2s'},
    domain={'x': [0.26, 0.49], 'y': [0, 1]}
))

fig.add_trace(go.Indicator(
    mode="number",
    value=profit_margin,
    title={"text": "Profit Margin"},
    number={'suffix': '%', 'valueformat': '.2f'},
    domain={'x': [0.52, 0.75], 'y': [0, 1]}
))

fig.add_trace(go.Indicator(
    mode="number",
    value=total_orders,
    title={"text": "Orders"},
    number={'valueformat': ',d'},
    domain={'x': [0.78, 1.00], 'y': [0, 1]}
))

fig.update_layout(
    height=250,
    title="Retail Sales KPI Dashboard"
)

fig.show()
```

Average order value(AOV)

```python
# Calculate AOV
total_revenue = df['Sales'].sum()
total_orders = df['OrderID'].nunique()

aov = total_revenue / total_orders

fig = go.Figure(go.Indicator(
    mode="gauge+number",
    value=aov,
    title={'text': "Average Order Value ($)"},
    number={'prefix': '$', 'valueformat': ',.2f'},
    gauge={
        'axis': {'range': [0, aov * 1.5]},
        'bar': {'color': "darkblue"},
        'steps': [
            {'range': [0, aov * 0.5], 'color': "#ffcccc"},
            {'range': [aov * 0.5, aov], 'color': "#fff3cd"},
            {'range': [aov, aov * 1.5], 'color': "#d4edda"}
        ]
    }
))

fig.update_layout(
    template="plotly_white",
    height=450
)

fig.show()
```

Return Rate and Repeat customer rate

```python
#return rate

returned_orders = (
    df[df['Returned'] == 'Yes']
    .shape[0]
)

total_orders = len(df)

return_rate = (
    returned_orders / total_orders
) * 100

print(f"Return Rate: {return_rate:.2f}%")
```

```python
#repeat customers rate

customer_orders = (
    df.groupby('CustomerID')['OrderID']
    .nunique()
)

repeat_customers = (
    customer_orders[customer_orders > 1]
    .count()
)

total_customers = (
    df['CustomerID']
    .nunique()
)

repeat_customer_rate = (
    repeat_customers / total_customers
) * 100

print(
    f"Repeat Customer Rate: {repeat_customer_rate:.2f}%"
)
```
